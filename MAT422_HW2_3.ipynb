```python
import numpy as np
from scipy.stats import alpha
import matplotlib.pyplot as plt
from random import sample
```

# **Alpha Continuous Distribution**



```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.alpha.html#scipy.stats.alpha is the link forr the example provided below

fig, ax = plt.subplots(1, 1)

# Calculate the first four moments
a = 3.57
mean, var, skew, kurt = alpha.stats(a, moments='mvsk')

# Display the pdf
x = np.linspace(alpha.ppf(0.01, a), alpha.ppf(0.99, a), 100)
ax.plot(x, alpha.pdf(x, a), 'r-', lw=5, alpha=0.6, label='alpha pdf')

# Freeze the distribution
rv = alpha(a)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')

# Generate random numbers
r = alpha.rvs(a, size=1000)

# plot data
ax.hist(r, density=True, bins='auto', histtype='stepfilled', alpha=0.2)
ax.set_xlim([x[0], x[-1]])
ax.legend(loc='best', frameon=False)
plt.show()
```

# **Random Sample**

Random samples are a randomly selected subset of a population. Random variables are said to form a simple random sample if all the X's are independent random variables and each X has the same probability distribution.

Python can be used to generate random samples using the random library and sample function.

```python
# Example of random samples using the random library

l1 = [1, 2, 3, 4]
string = "MathAndDataScience"
tuple1 = ("entry1", "entry2", "entry3", "entry4", "entry5")

print("Random with lists: ", sample(l1, 2))
print("Random with strings: ", sample(string, 5))
print("Random with tuples: ", sample(tuple1, 2))
```
