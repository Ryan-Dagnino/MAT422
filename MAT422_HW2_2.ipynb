```python
import numpy as np
from scipy.stats import poisson
from scipy.stats import binom
import matplotlib.pyplot as plt
```

# **Conditional Probability**

Some definitions:

P(A | B) = P(A ∩ B) / P(B)

P(A ∩ B) = P(A | B) ⋅ P(B)

P(Aᵢ_1 ∩ Aᵢ_2 ∩ … ∩ Aᵢₖ) = P(Aᵢ_1) ⋅ P(Aᵢ_2) ⋅ … ⋅ P(Aᵢₖ)

# **Discrete Random Variables**

Poisson distribution is a discrete distribution with the pmf of:

*p*(*x*;μ) = eᵘ ⋅ μˣ / x!

```python
# This is an example of the poisson distribution using the scipy.stats library which can be found at
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.poisson.html

fig, ax = plt.subplots(1,1)

mu = 0.6
mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')

M = np.arange(poisson.ppf(0.01, mu), poisson.ppf(0.99, mu))
ax.plot(M, poisson.pmf(M, mu), 'bo', ms=8, label='poisson pmf')
ax.vlines(M, 0, poisson.pmf(M, mu), colors='b', lw=5, alpha=0.5)
```
